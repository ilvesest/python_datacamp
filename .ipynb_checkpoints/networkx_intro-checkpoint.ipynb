##NetworkX API basics

```python
import networkx as nx

# initialize empty graph where we can add nodes and edges
G = nx.Graph()

# adding nodes
G.add_nodes_from([1,2,3])

# to see what nodes are in the graph
G.nodes()
```

```python
# adding edges to the graph
G.add_edge(1,2)

#seeing edges
G.edges()
```

Each tuple shows the nodes that are present in each edge. Metadata can be stored on graph as well. 

```python
# adding a label to node 1 with value 'blue'
G.node[1]['label'] = 'blue'

# retrive node information with metadata attached
G.nodes(data=True)
```

```python
# basic drawing functionality 
nx.draw(G)

import matplotlib.pyplot as plt

# this will show the node-link diagram rendering of the graph
plt.show()
```
