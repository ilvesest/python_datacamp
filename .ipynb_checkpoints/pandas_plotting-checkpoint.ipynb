# Plotting with pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'csv_files/aapl.csv'

aapl = pd.read_csv(file_path, index_col='Date', parse_dates=True)

aapl.head(6)
```

```python
close_arr = aapl['Close'].values

type(close_arr)
```

Few different options and different outputs of plotting data

## Plotting arrays (matplotlib)

```python
plt.plot(close_arr)

plt.show()

# on horizontal axis we have the date indexes of the array
```

## Plotting Series (matplotlib)

```python
close_series = aapl['Close']

plt.plot(close_series)

plt.show()

# on the horizontal axis we have the date_time 
```

## Plotting Series (pandas)

```python
close_series.plot()

plt.show()
```

## Plotting DataFrames (pandas)

```python
aapl.plot() # plots all Series at once

plt.show()
```

```python
plt.plot(aapl)  # plots all columns at once

plt.show()
```

```python
aapl.plot()

plt.yscale('log')

plt.show()
```

## Saving plots

```python
aapl.index
aapl.columns
```

```python
aapl.loc['2008':'2007', ['Open', 'High', 'Volume']].plot()

plt.savefig('aapl.png')

plt.savefig('aapl.pdf')

plt.show()
```

## Plot basics

```python
aapl.plot(color='red')

plt.title('Stocks')

plt.xlabel('Date')

plt.ylabel('Counts')
```

```python
aapl.columns
```

## Subplotting

```python
col_names = ['High', 'Low', 'Close']

aapl_3 = aapl[col_names]

aapl_3.plot(subplots=True)

plt.show()
```
