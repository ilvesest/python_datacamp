# Pivoting DF-s

For reshaping DFs

```python
import pandas as pd
id = [x for x in range(1, 5)]

dict = {'id': id, 
        'treatment' : ['A', 'A', 'B', 'B'], 
        'gender': ['F', 'M', 'F', 'M'], 
        'response': [5, 3, 8, 9]}
df = pd.DataFrame(dict)
df
```

pivot() method lets us choose which column to choose as index, and which to use as columns in the new DF. Values are set for the column which values are wanted to be shown.

```python
df.pivot(index='treatment',
         columns='gender',
         values='response')
```

By not declearing the values= value all the other columns are set as for the values

```python
df.pivot(index='treatment',
         columns='gender')
```

# Melting DFs

```python
import pandas as pd

id = [x for x in range(1, 5)]

dict = {'id': id, 
        'treatment' : ['A', 'A', 'B', 'B'], 
        'gender': ['F', 'M', 'F', 'M'], 
        'response': [5, 3, 8, 9]}
df = pd.DataFrame(dict)
df
```

```python
pivoted_df = df.pivot(index='treatment', columns='gender', values='response')
pivoted_df
```

Lets have a simiral DF but treatment not in the index

```python
dict_2 = {'treatment': ['A', 'B'], 'F': [5, 8], 'M': [3, 9]}

df2 = pd.DataFrame(dict_2)
df2
```

Melting is opposite operation to pivoting to convert pivoted DF into original shape.

```python
melted_only = pd.melt(df2)
melted_only
```

The melting method brings everything into a separate row. This is not what we wanted. This can be fixed if we specify the columns we'd like to keep.

### Specifying id_vars

```python
melted_var = pd.melt(df2, id_vars=['treatment'])
melted_var
```

We can specify which columns to turn to values

```python
melted_almost = pd.melt(df2, id_vars=['treatment'], value_vars=['F', 'M'])
melted_almost
```

We can change the column labels via:

```python
melted_correctly = pd.melt(df2, id_vars=['treatment'], value_vars=['F', 'M'],
                            var_name='gender', value_name='response')
melted_correctly
```
